- 아래 아스키 코드를 이용해 값을 구한다.
- 아스키 코드 값에 55를 뺀값.
- 10보다 작을 때는 숫자로 표현
- 각 자리에 해당 진법을 곱한다.

```python
ord('A'), ord('Z')
```

```python
Ns, B = input().split()
B = int(B)
ans = 0
Ns_l = len(Ns)-1

if B > 9 :
    for idx, N in enumerate(Ns) :
        num = ord(N)-55
        ans += num*(B**(Ns_l-idx))
else : 
    for idx, N in enumerate(Ns) :
        ans += int(N)*(B**(Ns_l-idx))
print(ans)
```

- B가 A to Z 값이 나오면 ord로 변경.
- 정답 X
- B가 A to Z 값이 나와도 문제에서 1~9 가 나와버리면 ord 값이 이상하게 나옴
    - ex) Z1Z2Z3 46
- 처음부터 나열하면서 비교하는것으로 변경

- 확인용 코드

```python
Ns, B = input().split()
B = int(B)
ans = 0
Ns_l = len(Ns)-1
for idx, N in enumerate(Ns):        # 자리수 만큼 제곱을 구하기 위해 enumerate 선언
    if ord(N) > 64 :                # ord 값이 65(A) 를 넘으면 ord에서 55를 뺀 값 (10 ~ 35) 를 구함
        num = ord(N)-55
        ans += num*(B**(Ns_l-idx))  # 숫자 x (해당 진수 ^ (자리 순서.)) 왼쪽부터 더해주기 위해 ((길이-1)-idx)로 진행.
        print(f'ord : {ord(N)}, num:{num}, ans:{ans}')
    else :                          # 정수값(0~9)가 나왔을때
        num = int(N)            
        ans += num*(B**(Ns_l-idx))
        print(f'num:{num}, ans:{ans}')
print(ans)
```

- 제출용 코드

```python
import sys
input = sys.stdin.readline
Ns, B = input().split()
B = int(B)
ans = 0
Ns_l = len(Ns)-1
for idx, N in enumerate(Ns):
    if ord(N) > 64 :
        num = ord(N)-55
        ans += num*(B**(Ns_l-idx))
    else :
        num = int(N)
        ans += num*(B**(Ns_l-idx))
print(ans)
```
