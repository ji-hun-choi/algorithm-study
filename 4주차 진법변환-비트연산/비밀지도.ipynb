```python
def bin_add(a1, a2, n): ## 2진번 계산까지.
    y = ''
    a = bin(a1)[2:]
    b = bin(a2)[2:]
    if a1 < a2:
        if len(a) < n or len(b) < n:
            a = '0'*(n-len(a)) + a
            b = '0'*(n-len(b)) + b
        for j in range(len(b)):
            if a[j] == '1' or b[j]== '1' :
                y += '1'
            else :
                y += '0'
    elif a1 > a2:
        b, a = a, b
        if len(a) < n or len(b) < n:
            a = '0'*(n-len(a)) + a
            b = '0'*(n-len(b)) + b
        for j in range(len(b)):
            if a[j] == '1' or b[j]== '1' :
                y += '1'
            else :
                y += '0'
    else :
        if len(a) < n or len(b) < n:
            y = '0'*(n-len(a))+a
        else :
            y = a
    return y

def change_shap(a):
    shap = ''
    for i in a:
        if i == '1':
            shap += '#'
        else :
            shap += ' '
    return shap

def solution(n, arr1, arr2):
    answer = []
    for i in range(len(arr1)):
        answer.append(change_shap(bin_add(arr1[i],arr2[i],n)))
    return answer
```

- 이진법 안에서도 계산이 가능한지 몰라 어렵게 구현을 하였다.

```python
def bin_add(a1, a2, n):
    ans = bin(a1|a2)[2:]
    l = len(bin(a1|a2)[2:])
    if n > l :
        ans='0' *(n-l) + ans
    return ans

def change_shap(ans):
    shap = ''
    for i in ans:
        if i == '1':
            shap += '#'
        else :
            shap += ' '
    return shap

def solution(n, arr1, arr2):
    answer = []
    for i in range(len(arr1)):
        answer.append(change_shap(bin_add(arr1[i],arr2[i],n)))
    return answer
```
