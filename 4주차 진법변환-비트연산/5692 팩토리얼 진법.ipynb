- 각 자리수의 숫자 곱하기 자릿수! 를 곱한다.
- 7 x 3! + 1 * 2! + 9 *  1!으로 진행 한다.
- 0이 나올 때까지 반복문을 실시한다.
- 문자로 입력을 받아 길이를 구한다.
- 1 ~ 길이+1 만큼 반복문 실시
- ans += quest[길이 - i] * fib(i)

```python
# 보여주기 코드
def fib(a): # 팩토리얼을 재귀로 받아냄.
    if a == 1 :
        return 1
    else :
        return  a * fib(a-1)

while True :
    quest = input()
    if quest == '0':
        break
    ans = 0
    q_len = len(quest)
    for i in range(1, q_len+1):
        ans += int(quest[q_len-i])*fib(i) # 뒤의 숫자 부터 진행
        print(f'quest[q_len-i] : {quest[q_len-i]}, fib(i) : {fib(i)}, ans = {ans}')
    print(ans)
```

```python
#제출용 코드
import sys

def fib(a):
    if a == 1 :
        return 1
    else :
        return  a * fib(a-1)

while True :
    # rstrip이 없으면 valueError가 나온다. 
    quest = sys.stdin.readline().rstrip() # 뒤 \n을 제거해주면 글을 받아옴
    if quest == '0':
        break
    ans = 0
    q_len = len(quest) # 문자의 길이를 받아온다.
    for i in range(1, q_len+1):
        ans += int(quest[q_len-i])*fib(i)
    print(ans)
```
