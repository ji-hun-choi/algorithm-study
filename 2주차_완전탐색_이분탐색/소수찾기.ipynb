## 조합 permutations 추가
- 아무리 시도해도 수의 조합을 만들기 불가능 하여서 수 조합을 구글링을 하였다.
- from itertools import permutations 을 임포트 하여서 진행을 했다.
- permutations(원하는문자, 갯수)


```python
from itertools import permutations

# 소수 판단 함수
def isPrime(number):
    if number == 0 : return False           # 결과 값이 0이 나와서 0도 추가
    elif number == 1 : return False         # 1일때 False
    k = int(number**0.5)
    for i in range(2, k+1):
        if number % i == 0 : return False
    return True


def solution(numbers):
    answer = 0
    lis1 = []
    for i in range(1, len(numbers)+1):      # 1 부터 numbers 길이만큼 반복
        for j in permutations(numbers, i):  # i의 길이 만큼 numbers의 조합을 구함.
            lis1.append(int(''.join(j)))    # 결과가 튜플이라 합친후 정수형으로 변환
    for i in set(lis1):                     # 중복 값 제거
        if isPrime(i):                      # 소수가 참일시 1 더한다.
            answer+=1
    return answer
```
