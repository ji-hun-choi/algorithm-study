## 문제 이해

- 모든 수포자의 찍는 방법은 반복적이다.
    - 1번 [1,2,3,4,5]
    - 2번 [2,1,2,3,2,4,2,5]
    - 3번 [3,3,1,1,2,2,4,4,5,5]
- 문제에서 맞추는 갯수가 많은 사람 순서대로 리스트 형식으로 출력된다.
- 브루트 포스 방식으로 진행

```python
def solution(answers):
    guess1 = [1,2,3,4,5] # 5
    guess2 =[2,1,2,3,2,4,2,5] # 8
    guess3 = [3,3,1,1,2,2,4,4,5,5] # 10

    answer = []
    lis = []

    lis.append(calculate(answers, guess1)) # 수포자 1의 문제 맞힌 갯수
    lis.append(calculate(answers, guess2)) # 수포자 2의 문제 맞힌 갯수
    lis.append(calculate(answers, guess3)) # 수포자 2의 문제 맞힌 갯수  

    for i, count in enumerate(lis): # 0 번째 부터 나열
        if count == max(lis): # 맞힌갯수 최대인 사람 인덱스 추가
            answer.append(i+1) # 인덱스는 0부터여서 i+1
    return answer

# 찍은게 얼마나 되는지 출력해주는 함수 이다.
def calculate(answers, guess):
    count = 0
    for i in range(len(answers)): # 문제 수 만큼 반복
        if answers[i] == guess[i%len(guess)]: # i % len(guess) 를 한 이유는 answer 보다 짧으며 길이에 다했을때 guess의 처음으로 돌아가기 위해서 이다.
            count += 1
    return count
```

```python
solution([1,3,2,4,2]), solution([1,2,3,4,5])
```
