    - 탐욕법

## 반복문

- 왼쪽부터 순서대로 검색

```python
def solution(trump):
    for i in range(len(trump)):
        if trump[i] == 8:
            return i
    return -1
```

## 재귀함수

```python
def solution(trump, loc):
    if trump[loc] == 8:
        return loc
    else :
        return solution(trump, loc+1)
```
