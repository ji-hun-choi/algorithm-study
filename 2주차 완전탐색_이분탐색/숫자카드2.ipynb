```python
import sys
rl = sys.stdin.readline
n = int(rl())
card = list(map(int, rl().split()))
m = int(rl())
deck = list(map(int, rl().split()))

# n = int(input())
# card = list(map(int, input().split()))
# m = int(input())
# deck = list(map(int, input().split())) 

answer = [0]*m

for i in range(m):
    answer[i] = card.count(deck[i])
print(*answer)
```

2.
- count 는 시간 초과가 나온다.
- for문 검색을 해본다
- 똑같이 시간 초과

```python
# import sys
# rl = sys.stdin.readline
# n = int(rl())
# card = list(map(int, rl().split()))
# m = int(rl())
# deck = list(map(int, rl().split()))

n = int(input())
card = list(map(int, input().split()))
m = int(input())
deck = list(map(int, input().split()))

answer = [0]*m

for i in range(m):
    cnt = 0
    for j in card:
        if j == deck[i]:
            cnt+=1
    answer[i] = cnt

print(*answer)
```

```python
# import sys
# rl = sys.stdin.readline
# n = int(rl())
# card = list(map(int, rl().split()))
# cards = {}
# for i in card:
#     if i in cards:
#         cards[i] += 1
#     else:
#         cards[i] = 1
# m = int(rl())
# deck = list(map(int, rl().split()))

n = int(input())
card = list(map(int, input().split()))
cards = {}
for i in card:
    if i in cards:
        cards[i] += 1
    else:
        cards[i] = 1
m = int(input())
deck = list(map(int, input().split()))

answer = [0]*m

for i in range(m):
    if deck[i] in cards:
        answer[i] = cards[deck[i]]

print(*answer)
```
